```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
#Média da coluna valor
dados.Valor.mean()
```

```python
# Pegando as médias segundo tipos de variaveis
# Vamos pegar as médias de acordo com os barrios:
# Copacabana, Barra de Tijuca, Ipanema, Leblon, Botafogo, Flamengo, Tijuca
```

```python
bairros = ["Barra da Tijuca", "Botafogo","Copacabana", "Ipanema", "Leblon", "Flamengo", "Tijuca"]
selecao = dados["Bairro"].isin(bairros)
dados = dados[selecao]
dados
```

```python
# só vou tirar os duplicados pra ver melhor o DF
dados.Bairro.drop_duplicates()
dados.head(10)
```

```python
grupo_bairro = dados.groupby('Bairro')
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, dados in grupo_bairro:
    print(bairro)
```

```python
for bairro, dados in grupo_bairro:
    print(type(dados))
```

```python
for bairro, dados in grupo_bairro:
  print(f'',{bairro}, {dados.Valor.mean()})
```

```python
# outra maneira
grupo_bairro.Valor.mean()
```

```python
# podemos colocar mais de um valor na operação
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
# retorna um dataframe com as médias de valor, condominio dos bairros selecionados!
```
