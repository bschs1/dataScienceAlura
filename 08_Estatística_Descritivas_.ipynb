# Estatísticas Descritivas I

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
#Média da coluna valor
dados.Valor.mean()
```

```python
# Pegando as médias segundo tipos de variaveis
# Vamos pegar as médias de acordo com os barrios:
# Copacabana, Barra de Tijuca, Ipanema, Leblon, Botafogo, Flamengo, Tijuca
```

```python
bairros = ["Barra da Tijuca", "Botafogo","Copacabana", "Ipanema", "Leblon", "Flamengo", "Tijuca"]
selecao = dados["Bairro"].isin(bairros)
dados = dados[selecao]
dados
```

```python
# só vou tirar os duplicados pra ver melhor o DF
dados.Bairro.drop_duplicates()
dados.head(10)
```

```python
grupo_bairro = dados.groupby('Bairro')
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print(bairro)
```

```python
for bairro, data in grupo_bairro:
    print(type(dados))
```

```python
for bairro, data in grupo_bairro:
  print(f'',{bairro}, {data.Valor.mean()})
```

```python
# outra maneira
grupo_bairro.Valor.mean()
```

```python
# podemos colocar mais de um valor na operação
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
# retorna um dataframe com as médias de valor, condominio dos bairros selecionados!
```

# Estatísticas Descritivas II

```python
grupo_bairro['Valor'].describe().round(2)
# Colunas do Describe:
# Frequencia, Media, Desvio Padrão, Valor Minimo, 25% é o primeiro 1/4, 50% é a mediana, 75% é o 3/4, e o max é o valor máximo
# Percebemos que o STD de Barra da Tijuca e Botafogo estão estranhos
# O max de botafogo e barra da tijuca tbm ta estranho
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Minimo', 'max': 'Máximo'})
# gerando outra estatísticas
```

```python
# vamos plotar!!!
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#plotando
plt.rc('figure', figsize = (20, 10))
fig = grupo_bairro['Valor'].std().plot.bar(color = 'Blue') 
#fig_mean = grupo_bairro['Valor'].mean().plot.bar(color = 'Blue') 
```

```python
#plots!!!
# esse plot é mt util pq aqui estamos so tirando a média mas poderíamos tirar varias outras infos
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'Blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize': 25})
fig.set_title('Valor da Média de Aluguel por Bairro', {'fontsize': 25})
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color = 'Blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize': 25})
fig.set_title('Valor da Média de Aluguel por Bairro', {'fontsize': 25})
```

```python
fig = grupo_bairro['Valor'].median().plot.bar(color = 'Blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize': 25})
fig.set_title('Valor da Média de Aluguel por Bairro', {'fontsize': 25})
```
