# Relatório VI
## Criando Novas Variáveis
### Uma varivavel de Valor Bruto, Valor de Aluguel por M², Variavel nova de tipo que é casa e apartamento

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Criando a Coluna " Valor Bruto "
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

```python
#valor m²
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados
```

```python
# tirando as casas decimais
dados['Valor m2'] = dados['Valor m2'].round(2)
dados
```

```python
#valor bruto m2
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
# Criando uma variável pra guardar casas e apartamentos
# usando lambda functions
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

```python
#Melhor forma, pq da pra deletar varias
#o método drop pode funcionar tanto na linha quanto na coluna e pra definir isso fazemos:
# axis = 1 = coluna
# axis = 0 = linha
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis=1, inplace=True)
```
