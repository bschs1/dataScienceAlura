## Agora nosso chefe quer apenas os tipos residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel.csv', sep=';')
dados.head(10)
```

```python
# vamos trabalhar com a coluna 'Tipo' dnv
# criando uma lista com a coluna 'Tipo' e tirando os duplicados
residencial = list(dados.Tipo.drop_duplicates()) 
residencial
```

```python
# Transformando em Lista e já filtrando para apenas o tipo residencial
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa Comercial',
 'Casa de Vila',]

residencial
```

```python
dados.head(10)
```

```python
#utilizando o método "isin"
selecao = dados.Tipo.isin(residencial) # passamos o parametro residencial
#o método isin serve para filtrar, como passamos residencial como parametro, ele vai retornar uma Series Booleana, comparando se o valor do dataframe está contido no parametro (residencial)
selecao.head(10) 
```

```python
# utilizamos o método isin para passarmos pro dataframe e ele seleciona, mantem no dataframe apenas os que retornaram true do isin
dados_residencial = dados[selecao] # criamos outra variável para criar um dataframe com o parametro residencial, o q retornou true de lá será substituído neste novo dataframe
dados_residencial
```

```python
 # só pra aparecer mais colunas
dados_residencial.head(10)
```

```python
# verificando pra ver se deu tudo certo
list(dados_residencial.Tipo.drop_duplicates())
```

```python
# vendo o tamanho do novo dataframe
dados_residencial.shape
# outro modo de ver o tamanho
len(dados_residencial)
```

```python
# comparando com o antigo df
dados.shape
```

```python
# o tanto que foi removido
len(dados) - len(dados_residencial)
```

```python
# recolocando o index
dados_residencial.head(10)
dados_residencial.index = range(len(dados_residencial))
```

```python
dados_residencial.head()
```

# Organizando dataFrames

```python
#criando um dataframe
data = [[1,2,3,], [4,5,6], [7,8,9]]
data
```

```python
list('321') # vai separar 1 por 1
```

```python
# passando uns parametros de lista so por passar mesmo pra fazer os exepmplos
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
# inplace precisa ser true pra mudar de vddd
df.sort_index(inplace=True) # sorting as colunas
df
```

```python
df.sort_index(inplace=True, axis=1) # do eixo 1
df
```

```python
# sort dos values
df.sort_values(by = 'X', inplace=True) # podemos usar o sort passando uma variavel utilizando o by = 'variavel'
df
```

```python
df.sort_values(by= '3', axis = 1, inplace= True) # dando sort utilizando a linha 3 
df
```

```python
df.sort_values(by = ['X', 'Y'], inplace= True) # pode colocar mais de uma variavel no by sort
df
```
